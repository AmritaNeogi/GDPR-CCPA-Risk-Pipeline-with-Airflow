```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 1) Load your processed daily counts:
# counts = (
#     pd.read_csv("data/processed/cleaned_policies.csv", parse_dates=["date"])
#       .groupby("date").size().reset_index(name="y")
#       .rename(columns={"date": "ds"})
# )


# Load actual counts
counts = (
    pd.read_csv("../data/processed/cleaned_policies.csv", parse_dates=["date"])
    .groupby("date").size().reset_index(name="y")
    .rename(columns={"date": "ds"})
)

# 2) Load latest forecast
forecast_path = sorted(glob.glob("../data/forecasts/forecast_*.csv"))[-1]
fc = pd.read_csv(forecast_path, parse_dates=["ds"])

# --- PLOT ---
fig, ax = plt.subplots(figsize=(14, 5))

# Actual values
ax.plot(counts["ds"], counts["y"], "o-", label="Actual daily count")

# Forecast
ax.plot(fc["ds"], fc["yhat"], "k--", label="Forecast")
ax.fill_between(fc["ds"], fc["yhat_lower"], fc["yhat_upper"], color="gray", alpha=0.3, label="Uncertainty")

# Vertical line: forecast start
ax.axvline(counts["ds"].max(), color="red", linestyle="--", label="Forecast Start")

# Smarter date ticks
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

# Labels & title
ax.set_title("GDPR Policy Updates — Daily Actual vs. Forecast")
ax.set_xlabel("Date")
ax.set_ylabel("Policy Count")
ax.legend(loc="upper left")

plt.tight_layout()
plt.show()
```


## 📊 **What This Graph Represents**

### **Title:**

**“GDPR Policy Updates — Daily Actual vs. Forecast”**
→ We're comparing the real number of GDPR-related policy announcements per day with a model's prediction of how that might continue.

---

### 🧭 **Key Elements in the Graph**

| Element                      | Meaning                                                                                                                        |
| ---------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| **Blue dots + line**         | The actual number of policy updates posted on each day (historical data from EDPB, scraped and processed by your pipeline).    |
| **Dashed black line**        | The predicted daily count of GDPR-related updates based on a Prophet forecast model (`yhat`).                                  |
| **Gray shaded area**         | The 95% confidence interval or **uncertainty band**. Prophet estimates where the true values might reasonably fall.            |
| **Red vertical dotted line** | The point in time where **forecasting begins**. Everything before this is historical data. After this, it’s model predictions. |

---

## 🔍 **How to Interpret It**

* We have **daily resolution** here — every point is one day of activity.
* We can see clear **spikes** on certain dates — possibly press releases, enforcement decisions, or guidelines dropped.
* The model learns a relatively **stable average daily count** (around \~1–2/day).
* **Prediction after the red line**:

  * The model continues this trend into the future.
  * It doesn’t predict the spikes well (which is normal — Prophet isn’t good at rare shocks).
  * The uncertainty band **widens** — confidence in predictions goes down further into the future.

---

## 📌 Why This Is Useful

* Helps monitor **regulatory intensity over time**
* You can create alerts if **real activity exceeds upper bound**
* Forecasted trends support **resourcing & compliance planning**

---

## Flag Anomalies

Highlight days where actual updates exceed the forecast's upper confidence limit.

```python
# Merge actual + forecast
merged = pd.merge(counts, fc[["ds", "yhat", "yhat_upper"]], on="ds", how="inner")

# Flag days where actual count exceeds forecast upper bound
merged["anomaly"] = merged["y"] > merged["yhat_upper"]

# Plot with anomaly markers
fig, ax = plt.subplots(figsize=(14, 5))

# All points
ax.plot(merged["ds"], merged["y"], "o-", label="Actual daily count")

# Anomalies
ax.plot(merged[merged["anomaly"]]["ds"],
        merged[merged["anomaly"]]["y"],
        "ro", label="Anomalies")

# Forecast
ax.plot(merged["ds"], merged["yhat"], "k--", label="Forecast")
ax.fill_between(merged["ds"], merged["yhat"], merged["yhat_upper"],
                color="gray", alpha=0.3, label="Uncertainty")

ax.axvline(counts["ds"].max(), color="red", linestyle="--", label="Forecast Start")

# Labels
ax.set_title("Anomaly Detection — Policy Spikes")
ax.set_xlabel("Date")
ax.set_ylabel("Policy Count")
ax.legend(loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Switch to Weekly Resolution

```python
counts = (
    pd.read_csv("../data/processed/cleaned_policies.csv", parse_dates=["date"])
    .groupby(pd.Grouper(key="date", freq="W-MON"))  # weekly, starting Monday
    .size()
    .reset_index(name="y")
    .rename(columns={"date": "ds"})
)
```
